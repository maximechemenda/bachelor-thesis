# GPT

```python
TITLE_FONT_SIZE = 12
AXIS_FONT_SIZE = 11

VMIN = -1.89
VMAX = 6.87
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Define the data
data = [[+0.67, +0.30, +0.22, +1.25, +0.30, +1.18, -0.88, -0.40, -0.77],
  [-0.34, -0.30, -0.17, +0.62, +0.30, +0.46, -1.40, -0.30, -0.81],
  [+2.42, +0.7, +1.67, +0.84, +0.3, +0.34, +0.18, +0.7, +0.64],
  [+0.72, -0.4, +0.71, +1.38, -0.1, +1.18, +0.1, 0, +0.1],
  [-0.23, 0, -0.34, -0.75, -0.1, -0.51, +0.1, +0.1, +0.08],
  [+5.28, +2.3, +4.77, +5.6, +0.9, +5.0, +6.87, +3.3, +6.5]
]

columns = ['Pro. R-1', 'Pro. R-2', 'Pro. R-L', 'Aca. R-1', 'Aca. R-2', 'Aca. R-L', 'Com. R-1', 'Com. R-2', 'Com. R-L']
index = ['Pro. (S)', 'Aca. (S)', 'Com. (S)', 'Pro. (R)', 'Aca. (R)', 'Com. (R)']


df = pd.DataFrame(data, columns=columns, index=index)

fig, ax = plt.subplots(figsize=(7, 7))


heatmap = sns.heatmap(df, annot=True, fmt='.2f', vmin=VMIN, vmax=VMAX)
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=45)


heatmap.set_title('Variation of ROUGE scores after performing paraphrasing with GPT', fontsize=TITLE_FONT_SIZE)


plt.xlabel('Test domain and metric', fontsize=AXIS_FONT_SIZE)
plt.ylabel('Train domain and model', fontsize=AXIS_FONT_SIZE)

# Display the plot
# plt.show()
plt.savefig('heatmap-gpt.png', dpi=300, bbox_inches='tight')

```
