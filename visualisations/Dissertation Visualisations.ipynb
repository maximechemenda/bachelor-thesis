# Train on product (summarizer & retriever)

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
colors = {
    "academic_orig": "tab:blue",
    "academic_aug": "tab:orange",
    "product_orig": "tab:green",
    "product_aug": "tab:purple",
    "committee_orig": "red",
    "committee_aug": "gray",
}
```

```python
import matplotlib as mpl
mpl.rcParams['font.size'] = 14

# Common setup

fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(14,5))

# Labels for x-axis
labels = ['R-1', 'R-2', 'R-L']

# Width of bars
bar_width = 0.15

# Setup per domain
out_of_domain_labels = ['Aca. (orig.)', 'Aca. (aug.)', 'Com. (orig.)', 'Com. (aug.)']

# ---------- SUMMARIZER --------
# ---------- SUMMARIZER --------
# ---------- SUMMARIZER --------
# ---------- SUMMARIZER --------

orig = [(34.63, 10.3, 30.89), (25.59, 5.2, 22.48,), (31.77, 7.2, 27.87)]
aug = [(35.30, 10.6, 31.11), (26.84, 5.5, 23.66), (30.89, 6.8, 27.15)]

# Output magintudes for original
zero_point_orig = orig[0]
zero_point_aug = aug[0]

diff_orig = [(abs(x[0]-zero_point_orig[0]), abs(x[1]-zero_point_orig[1]), abs(x[2]-zero_point_orig[2])) for x in orig]
diff_aug = [(abs(x[0]-zero_point_aug[0]), abs(x[1]-zero_point_aug[1]), abs(x[2]-zero_point_aug[2])) for x in aug]

# Remove product
diff_orig = diff_orig[1:]
diff_aug = diff_aug[1:]

# THESE ARE THE VALUES I AM USING
academic_orig_summarizer = list(diff_orig[0])
committee_orig_summarizer = list(diff_orig[1])
academic_aug_summarizer = list(diff_aug[0])
committee_aug_summarizer = list(diff_aug[1])

# Positions of bars on x-axis
r1 = np.arange(len(labels))
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width + 0.05 for x in r2]
r4 = [x + bar_width for x in r3]

ax1.bar(r1, academic_orig_summarizer, color=colors['academic_orig'], width=bar_width, edgecolor='white', label=out_of_domain_labels[0])
ax1.bar(r2, academic_aug_summarizer, color=colors['academic_aug'], width=bar_width, edgecolor='white', label=out_of_domain_labels[1])
ax1.bar(r3, committee_orig_summarizer, color=colors['committee_orig'], width=bar_width, edgecolor='white', label=out_of_domain_labels[2])
ax1.bar(r4, committee_aug_summarizer, color=colors['committee_aug'], width=bar_width, edgecolor='white', label=out_of_domain_labels[3])




# Adding labels, title, and legend
ax1.set_xlabel('Metric')
ax1.set_ylabel('Out-of-domain performance gap')
ax1.set_xticks([r + 2*bar_width for r in range(len(labels))])
ax1.set_xticklabels(labels)

# ax1.set_title('Effect on the out-of-domain gap after augmenting data with paraphrasing (Train set: Product')
ax1.set_title('Summariser')
ax1.legend()


# ---------- RETRIEVER --------
# ---------- RETRIEVER --------
# ---------- RETRIEVER --------
# ---------- RETRIEVER --------

orig = [(20.30, 5.3, 18.41), (13.39, 2.1, 12.05), (6.5, 3.2, 6.3)]
aug = [(21.02, 4.9, 19.12), (14.77, 2.0, 13.23), (6.6, 3.2, 6.4)]

# Output magintudes for original
zero_point_orig = orig[0]
zero_point_aug = aug[0]

diff_orig = [(abs(x[0]-zero_point_orig[0]), abs(x[1]-zero_point_orig[1]), abs(x[2]-zero_point_orig[2])) for x in orig]
diff_aug = [(abs(x[0]-zero_point_aug[0]), abs(x[1]-zero_point_aug[1]), abs(x[2]-zero_point_aug[2])) for x in aug]

# Remove product
diff_orig = diff_orig[1:]
diff_aug = diff_aug[1:]

# THESE ARE THE VALUES I AM USING
academic_orig_retriever = list(diff_orig[0])
committee_orig_retriever = list(diff_orig[1])
academic_aug_retriever = list(diff_aug[0])
committee_aug_retriever = list(diff_aug[1])

# Positions of bars on x-axis
r1 = np.arange(len(labels))
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width + 0.05 for x in r2]
r4 = [x + bar_width for x in r3]

ax2.bar(r1, academic_orig_retriever, color=colors['academic_orig'], width=bar_width, edgecolor='white', label=out_of_domain_labels[0])
ax2.bar(r2, academic_aug_retriever, color=colors['academic_aug'], width=bar_width, edgecolor='white', label=out_of_domain_labels[1])
ax2.bar(r3, committee_orig_retriever, color=colors['committee_orig'], width=bar_width, edgecolor='white', label=out_of_domain_labels[2])
ax2.bar(r4, committee_aug_retriever, color=colors['committee_aug'], width=bar_width, edgecolor='white', label=out_of_domain_labels[3])


# Adding labels, title, and legend
ax2.set_xlabel('Metric')
ax2.set_ylabel('Out-of-domain performance gap')
ax2.set_xticks([r + 2*bar_width for r in range(len(labels))])
ax2.set_xticklabels(labels)

ax2.set_title('Retriever')
ax2.legend()

# Displaying plot
plt.savefig('bar-charts-train-product-gpt.png', dpi=300, bbox_inches='tight')
# plt.show()
```
